```
from pyspark.sql.functions import *
import urllib
```

```
file_type = "csv"
first_row_is_header = "true"
delimiter = ","

aws_key_df = spark.read.format(file_type)\
 .option("header", first_row_is_header)\
 .option("sep", delimiter)\
 .load("/FileStore/tables/aws_key.csv")
```

```
ACCESS_KEY = aws_key_df.where(col('User name')=='Databricks').select('Access key ID').collect()[0]['Access key ID']
SECRET_KEY = aws_key_df.where(col('User name')=='Databricks').select('Secret access key').collect()[0]['Secret access key']
```

```
ENCODED_SECRET_KEY = urllib.parse.quote(string=SECRET_KEY, safe="")
```

```
AWS_S3_BUCKET = "yahoo-finances-bg"

MOUNT_NAME = "/mnt/yahoo-finances-bg"

SOURCE_URL = "s3n://{0}:{1}@{2}".format(ACCESS_KEY, ENCODED_SECRET_KEY, AWS_S3_BUCKET)

dbutils.fs.mount(SOURCE_URL, MOUNT_NAME)
```
